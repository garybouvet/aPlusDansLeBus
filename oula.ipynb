```python
import requests
import pandas as pd
import time
import schedule
api_key = '9JVETI8PUF'
api_url = f"https://data.bordeaux-metropole.fr/geojson?key={api_key}&typename=ci_vcub_p"
# Define an empty dataframe to store the results
concatenated_df = pd.DataFrame()
def fetch_data():
    try:
        # Send a GET request to the API
        response = requests.get(api_url)
        # Check if the request was successful
        response.raise_for_status()
        # Extract the JSON data from the response
        data = response.json()
        # Extract the records from the JSON
        records = data["features"]
        # Extract the relevant data from each record and store them in a list of dictionaries
        extracted_data = []
        for record in records:
            extracted_record = {
                "type": record["type"],
                "geometry": record["geometry"],
                **record["properties"]
            }
            extracted_data.append(extracted_record)
        # Convert the extracted data to a dataframe
        df = pd.DataFrame(extracted_data)
        # Extract the coordinates from the "geometry" column and store them in separate columns
        df["longitude"] = df["geometry"].apply(lambda x: x["coordinates"][0] if x else None)
        df["latitude"] = df["geometry"].apply(lambda x: x["coordinates"][1] if x else None)
        # Drop the original "geometry" column
        df.drop("geometry", axis=1, inplace=True)
        # Concatenate the current dataframe with the previous results
        global concatenated_df
        concatenated_df = pd.concat([concatenated_df, df], ignore_index=False)
        # Save the concatenated dataframe to a CSV file
        concatenated_df.to_csv('data.csv', index=False)
        # Print the concatenated dataframe
        print(concatenated_df)
    except requests.exceptions.RequestException as err:
        print("Request failed:", err)
    except (KeyError, ValueError) as err:
        print("Error occurred while processing the response:", err)
# Schedule the data fetching every 10 minutes
schedule.every(1).minutes.do(fetch_data)
# Run the scheduler indefinitely
while True:
    schedule.run_pending()
    time.sleep(1) 
```

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# Lire le fichier CSV
df = pd.read_csv('/Users/garybouvet/Desktop/Streamlit_projet_3/gdfbustrambat.csv')

# Convertir la colonne 'geometry' en objets géométriques
df['geometry'] = df['geometry'].apply(wkt.loads)

# Créer un GeoDataFrame à partir du DataFrame
gdf = gpd.GeoDataFrame(df, geometry='geometry')

# Simplifier les objets géométriques
gdf['geometry_simplified'] = gdf['geometry'].simplify(tolerance=0.001)

# Convertir les objets géométriques simplifiés en WKT (Well-Known Text)
df['geometry_simplified'] = gdf['geometry_simplified'].apply(lambda geom: geom.wkt)

# Enregistrer le DataFrame dans un nouveau fichier CSV
df.to_csv('/Users/garybouvet/Desktop/Streamlit_projet_3/df_simplified.csv', index=False)
```
